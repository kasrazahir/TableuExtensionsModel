```python
from sklearn.datasets.samples_generator import make_blobs
from sklearn.ensemble import IsolationForest
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
z, y = make_blobs(n_samples=500, centers=2,cluster_std=0.9)
x1 = [i[0] for i in z]
x2 = [i[1] for i in z]
```

```python
# plt.scatter(x1,x2,c=y, cmap="coolwarm")
# 
```

```python
df = pd.DataFrame({'x':x1,'y':x2, 'cluster':1})
df.to_csv('../scatter.csv')
```

```python
# plt.scatter(df['x'], df['y'], c=df['cluster'], cmap='coolwarm')
```

```python
df['x']=pd.to_numeric(df['x'])
sns_plot = sns.lmplot('x','y',df, fit_reg=False, hue="cluster", legend=False)
sns_plot.savefig("output.png")
```

```python
data = df
fig = plt.figure()
print('data in graph', data.head())
plt.scatter(data['x'], data['y'], c=data['cluster'], cmap='coolwarm')
# plt.xticks([])
# plt.yticks([])
fig.savefig('22.png')
```

```python
# dfr = df.values.reshape(-1,2)
IF = IsolationForest()
IF.fit(df)
ifpred = IF.predict(df)
```

```python
df['cluster'] = (df['cluster']+1)*ifpred
```

```python
plt.scatter(df['x'], df['y'], c=df['cluster'], cmap='coolwarm')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)
```

```python
kpred = kmeans.fit_predict(df[['x','y']])
```

```python
df['cluster']=kpred
```
